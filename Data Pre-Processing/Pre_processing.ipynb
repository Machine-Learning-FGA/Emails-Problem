### Bag of Words
Processando os dados com a técnica chamada _Bag of words_ ou punhado de palavras, em tradução livre.

Isso deve levar algum tempo, por volta de ~20 min.

```python
import os
import codecs
from time import time
import re
import nltk
# nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

path = '../dataset'
data = [['corpus', 'label']]
count_success = 0
count_fail = 0
start_time = time()
for root, dirs, files in os.walk(path):
    for file in files:
        try:
            file_path = os.path.join(root, file) # '../dataset/alt.atheism/51121'
            f = open(file_path, 'r', encoding='ISO-8859-1')
            head, content = f.read().split('\n', 1) # removing "from" to avoid overfitting
            content = re.sub('[^a-zA-Z]', ' ', content)
            content = content.lower()
            content = content.split()
            ps = PorterStemmer()
            content = [ps.stem(word) for word in content if not word in set(stopwords.words('english'))]
            content = ' '.join(content)

            row = []
            row.append(content)
            label = root.split(os.path.sep)[-1]
            row.append(label)
            data.append(row)
            print('Succes: ', file_path)
            count_success += 1
        except Exception as e:
            print('Failed at: ', file_path)
            print('Error', e)
            count_fail += 1
            pass


import csv 

out_file_name = '../20newgroups_bow.csv'
with open(out_file_name, 'w') as out_file:
    writer = csv.writer(out_file)
    writer.writerows(data) # Write to file
    print('File parsed. Created file %s!' % out_file_name)

duration = time() - start_time
print('Finished with %0.3fs. Parsed %d documents and did not parsed %d documents.' % (duration, count_success, count_fail))
```
